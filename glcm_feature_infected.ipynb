```python
import numpy as np
import matplotlib as plt
import glob
import cv2
import os
import seaborn as sns
import pandas as pd
from skimage.filters import sobel
from skimage.feature import greycomatrix,greycoprops
from skimage.measure import shannon_entropy
import csv



imgs2 = []
labels2 = []

for img2 in glob.glob(r"C:\Users\91966\Downloads\archive\TB_Chest_Radiography_Database\Tuberculosis\*.png"):
            base2 = os.path.basename(img2)
            img2 = cv2.imread(img2)
            gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)           
            resize = cv2.resize(gray2,(200,200))
           
            matrix_coocurrence2 = greycomatrix(resize, [1], [0, np.pi/4, np.pi/2, 3*np.pi/4], levels=256, normed=False, symmetric=False)
            
            imgs2.append(resize)
            labels2.append(base2)
            
            
            
def calc_glcm_all_agls_2(img, label, props, dists=[5], agls=[0, np.pi/4, np.pi/2, 3*np.pi/4], lvl=256, sym=True, norm=True):
    
    glcm2 = greycomatrix(img, 
                        distances=dists, 
                        angles=agls, 
                        levels=lvl,
                        symmetric=sym, 
                        normed=norm)
    feature2 = []
    glcm_props2 = [propery for name in props for propery in greycoprops(glcm2, name)[0]]
    for item2 in glcm_props2:
            feature2.append(item2)
    feature2.append(label2) 
    
    #feature['Tuberculosis']=1
    
    return feature2



properties = ['dissimilarity', 'correlation', 'homogeneity', 'contrast', 'ASM', 'energy']

glcm_all_agls_2 = []
for img2, label2 in zip(imgs2, labels2): 
    glcm_all_agls_2.append(
            calc_glcm_all_agls_2(img2, 
                                label2, 
                                props=properties)
                            )
columns = []
angles2 = ['0', '45', '90','135']
for name2 in properties :
    for ang2 in angles2:
        columns.append(name2 + "_" + ang2)
        
columns.append("label2")


glcm_df2 = pd.DataFrame(glcm_all_agls_2, 
                      columns = columns)

glcm_df2.to_csv(r"C:\Users\91966\Downloads\csv2.csv")



#glcm_df.head(15)
aa2 = pd.read_csv(r"C:\Users\91966\Downloads\csv2.csv")
aa2.insert(2, column = "tuberculosis", value =1)
aa2.head()
aa2.to_csv(r"C:\Users\91966\Downloads\aa2.csv")
```
